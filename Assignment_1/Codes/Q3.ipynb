```python
from BP_reward import get_reward
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from statistics import NormalDist
import scipy.stats
from doctors import *
```

# Inputs

```python
s_id = 810100476
drugs = [1, 2]
```

# Confidence Interval

```python
def confidence_interval(size, doctor):
    t_score = np.abs(scipy.stats.t.ppf(q=0.05/2, df=size-1))
    rwrds = np.zeros([size,100])
    for i in range(size):
        rewards_mean, _ = doctor(s_id, drugs)
        rwrds[i,:] = rewards_mean
    rwrds_mean, rwrds_std = rwrds.mean(axis=0), rwrds.std(axis=0)
    ci_up, ci_down = rwrds_mean + t_score*rwrds_std/np.sqrt(size), rwrds_mean - t_score*rwrds_std/np.sqrt(size)
    return ci_up, ci_down, rwrds_mean
```

# Plots


### size = 5

```python
size = 5
plt.subplots(3,1, figsize=(20,18), sharex=True, sharey=True)

plt.subplot(311)
ci_up, ci_down, rwrds_mean = confidence_interval(size, doctor_A)
plt.plot(range(1,101), rwrds_mean, label='win-stay lose-shift')
plt.fill_between(range(1,101), ci_up, ci_down, alpha=0.2)
plt.ylabel('Reward', fontsize=15, labelpad=10)
plt.legend(fontsize=30)
plt.grid()

plt.subplot(312)
ci_up, ci_down, rwrds_mean = confidence_interval(size, doctor_B)
plt.plot(range(1,101), rwrds_mean, label='fully random')
plt.fill_between(range(1,101), ci_up, ci_down, alpha=0.2)
plt.ylabel('Reward', fontsize=15, labelpad=10)
plt.legend(fontsize=30)
plt.grid()

plt.subplot(313)
ci_up, ci_down, rwrds_mean = confidence_interval(size, doctor_C)
plt.plot(range(1,101), rwrds_mean, label='doctor C')
plt.fill_between(range(1,101), ci_up, ci_down, alpha=0.2)
plt.xlabel('Trial', fontsize=15, labelpad=10)
plt.ylabel('Reward', fontsize=15, labelpad=10)
plt.legend(fontsize=30)
plt.grid()

plt.xticks(ticks=[1,20,40,60,80,100], labels=[1,20,40,60,80,100])
plt.suptitle('Reward per Trial (with Confidence Intervals)(size=5)', y=0.99, fontsize=35)
plt.tight_layout(pad=1)
```
